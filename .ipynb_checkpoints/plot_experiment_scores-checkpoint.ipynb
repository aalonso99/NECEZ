```python
import numpy as np
# import seaborn as sns
import matplotlib.pyplot as plt
import itertools
import os
```

```python
algorithms = ['EZ', 'NECEZ']
consistencies = ['consistency_05', 'consistency_1', 'consistency_2', 'consistency_5']
# consistencies = ['consistency_05']
```

### Plot training scores

```python
X_SIZE_EZ = 10_000
X_EZ = np.arange(X_SIZE_EZ)

X_SIZE_NECEZ = 1_600
X_NECEZ = np.arange(X_SIZE_NECEZ)
```

```python
scores = { f"{algo}_{consistency}":
           [ np.asarray(
             [ float(score) 
               for score in
               open(os.path.join(f"./runs/{algo}/{consistency}", 
                                 run_log_dir, 
                                 'scores.log'), 
                    'rb').read().splitlines() 
             ] )
             for run_log_dir in os.listdir(f"./runs/{algo}/{consistency}") ] 
          for algo, consistency in itertools.product(algorithms, consistencies)
          if os.path.exists(f"./runs/{algo}") and os.path.exists(f"./runs/{algo}/{consistency}")
         }
```

```python
run_x_size = { f"{algo}_{consistency}" : X_SIZE_EZ if algo=='EZ' else X_SIZE_NECEZ
               for algo, consistency in itertools.product(algorithms, consistencies)
               if os.path.exists(f"./runs/{algo}") and os.path.exists(f"./runs/{algo}/{consistency}")
             }

resized_scores_arrays = { key : np.asarray(
                                [ np.interp(np.linspace(0, run_scores.size - 1, run_x_size[key]), 
                                            np.arange(run_scores.size), 
                                            run_scores) 
                                  for run_scores in values
                                ] )
                          for key, values in scores.items()
                        }
```

```python
run_x = { key : np.arange(size) for key, size in run_x_size.items() }
COLORS = [ (1,0.3,0.3), (0.2,0.8,0.1), (0.7, 0.8, 0.95), (0.5,0.5,0.5) ]
# COLORS = [ np.random.rand(3) for _ in len(consistencies) ]

for algo in algorithms:
    fig, ax = plt.subplots(figsize=(8,8))
    for consistency, color in zip(consistencies, COLORS):
        experiment_name = f"{algo}_{consistency}"
        if experiment_name in resized_scores_arrays:
            experiment_scores = resized_scores_arrays[experiment_name]
            scores_mean = experiment_scores.mean(axis=0)
            scores_std = experiment_scores.std(axis=0)
            
            ecolor = np.hstack([color, 0.01])

            ax.errorbar(run_x[experiment_name], scores_mean, yerr=scores_std, 
                        color=color, ecolor=ecolor, label=consistency)
        else:
            continue
    ax.set_title(f'Evolución del score de entrenamiento ({algo})')
    ax.set_ylabel('Score')
    ax.set_xlabel('Frames de entrenamiento')
    ax.legend()
    plt.show()
```

### Validation

```python
import pickle
import pandas as pd
from IPython.display import display

def show_full_results(validation_scores):
    val_scores_tables = { algo_consistency : pd.DataFrame(val_score)
                          for algo_consistency,val_score in validation_scores.items() }
    for algo_consistency in val_scores_tables:
        print(algo_consistency)
        val_scores_tables[algo_consistency].columns = [f"Agente {i}" for i 
                                                       in range(len(val_scores_tables[algo_consistency].columns))]
        val_scores_tables[algo_consistency].index = [f"Run {i}" for i 
                                                       in range(len(val_scores_tables[algo_consistency]))]
        val_scores_tables[algo_consistency].loc['Media'] = val_scores_tables[algo_consistency].mean()
        display(val_scores_tables[algo_consistency])
        
def show_stats(validation_scores):
    val_scores_tables = { algo_consistency : pd.DataFrame(val_score)
                          for algo_consistency,val_score in validation_scores.items() }
    for algo_consistency in val_scores_tables:
        print(algo_consistency)
        val_scores_tables[algo_consistency].columns = [f"Agente {i}" for i 
                                                       in range(len(val_scores_tables[algo_consistency].columns))]
        val_scores_tables[algo_consistency].index = [f"Run {i}" for i 
                                                       in range(len(val_scores_tables[algo_consistency]))]
        stats_df = val_scores_tables[algo_consistency].describe()
        stats_df.drop(index=["count", "25%", "75%"], inplace=True)
        stats_df.rename({"mean":"Media", "std":"σ", "min":"Mín.", "50%":"Mediana" "max":"Máx."}, inplace=True)
        
        display(stats_df.round(decimals=2))
```

```python
validation_scores = pickle.load(open("validation_scores_EZ.pkl", 'rb'))
show_stats(validation_scores)
```

```python
validation_scores = pickle.load(open("validation_scores_EZ.pkl", 'rb'))
show_full_results(validation_scores)
```

```python
validation_scores = pickle.load(open("validation_scores_NECEZ.pkl", 'rb'))
show_full_results(validation_scores)
```
